## **YOLO v10**

```python
model_2 = YOLO('yolov10n.pt')
```

```python
model_2.train(
    data='/content/proyecto_IA_v2/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    project='runs/yolov10_runs',
    name='exp_custom',
    exist_ok=False
)
```
